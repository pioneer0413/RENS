```python
import tonic

nmnist = tonic.datasets.NMNIST("/DATA/hwkang", train=False)
events, label = nmnist[0]
```

```python
transform_n = tonic.transforms.Compose(
    [
        tonic.transforms.ToFrame(
            sensor_size=nmnist.sensor_size,
            time_window=10000,
        ),
    ]
)

transform_j = tonic.transforms.Compose(
    [
        tonic.transforms.TimeJitter(std=10000, clip_negative=True),
        tonic.transforms.ToFrame(
            sensor_size=nmnist.sensor_size,
            time_window=10000,
        ),
    ]
)

frames_n = transform_n(events)
frames_j = transform_j(events)
```

```python
ani_n = tonic.utils.plot_animation(frames_n)
ani_j = tonic.utils.plot_animation(frames_j)
```
