```python
import tonic

nmnist = tonic.datasets.NMNIST("/DATA/hwkang", train=False)
events, label = nmnist[0]
```

```python

transform1 = tonic.transforms.Compose(
    [
        tonic.transforms.DropEvent(p=1 / 3),
        tonic.transforms.ToFrame(
            sensor_size=nmnist.sensor_size,
            time_window=10000,
        ),
    ]
)

transform2 = tonic.transforms.Compose(
    [
        tonic.transforms.DropEvent(p=2 / 3),
        tonic.transforms.ToFrame(
            sensor_size=nmnist.sensor_size,
            time_window=10000,
        ),
    ]
)

transform3 = tonic.transforms.Compose(
    [
        tonic.transforms.DropEvent(p=3 / 3),
        tonic.transforms.ToFrame(
            sensor_size=nmnist.sensor_size,
            time_window=10000,
        ),
    ]
)
```

```python
frames1 = transform1(events)
frames2 = transform2(events)
frames3 = transform3(events)
```

```python
ani = tonic.utils.plot_animation(frames1)
# Display the animation inline in a Jupyter notebook
from IPython.display import HTML
HTML(ani.to_jshtml())
```
