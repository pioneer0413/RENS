```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
categories = ['Default', 'Noise Trained', 'SpikingCNN']
noise = [0, 0.455, 0]
infer = [0.775, 0.857, 28.416]
backprop = [2.297, 2.681, 25.851]
```

```python
plt.figure(figsize=(8,6))

#plt.bar(categories, batch, label='Batch Time')
plt.bar(categories, noise, bottom=[i+j for i,j in zip(infer, backprop)], label='Noise Generation Time', color='orange')
plt.bar(categories, infer, bottom=backprop, label='Inference Time', color='green')
plt.bar(categories, backprop, label='Backpropagation Time', color='blue')


plt.ylabel('Latency (ms)')

plt.legend()

plt.savefig('CNN_vs_SpikingCNN_Latency.png')
plt.show()
```
