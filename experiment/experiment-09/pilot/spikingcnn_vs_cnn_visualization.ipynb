```python
import matplotlib.pyplot as plt
```

```python
def visualize_snn_vs_cnn(li1, li2, path):
    # x축에 해당하는 값 (인덱스 또는 다른 값)
    xticks=['20', '25', '30', '35']

    # 그래프 그리기
    plt.figure(figsize=(8, 5))  # 그래프 크기 설정
    plt.plot(xticks, li1, label='SpikingCNN', marker='o', markersize=12, color='#ea801c', linewidth=3)  # 첫 번째 리스트의 그래프
    plt.plot(xticks, li2, label='CNN', marker='s', markersize=12, color='#1a80bb', linewidth=3)  # 두 번째 리스트의 그래프

    # 그래프 제목 및 레이블 설정
    plt.xlabel('PSNR (dB)', fontsize=16)
    plt.ylabel('Accuracy (%)', fontsize=16)

    plt.tick_params(axis='x', labelsize=14)  # x축 tick 크기 설정
    plt.tick_params(axis='y', labelsize=14)  # y축 tick 크기 설정

    # 범례 추가
    plt.legend(fontsize=16)

    plt.savefig(path)

    # 그래프 출력
    plt.show()
```

```python
scnn_uniform = [9.74, 22.06, 89.82, 98.39]
cnn_uniform = [10.32, 14.82, 32.38, 86.88]

visualize_snn_vs_cnn(scnn_uniform, cnn_uniform, 'spikingcnn_vs_cnn_uniform.png')
```

```python
scnn_jitter = [90.01, 98.00, 98.41, 98.51]  # 첫 번째 리스트
cnn_jitter = [78.24, 82.82, 87.75, 88.43]  # 두 번째 리스트

visualize_snn_vs_cnn(scnn_jitter, cnn_jitter, 'spikingcnn_vs_cnn_jitter.png')
```

```python
scnn_compound = [55.02, 87.72, 95.29, 98.42]  # 첫 번째 리스트
cnn_compound = [22.58, 48.64, 51.91, 86.97]  # 두 번째 리스트

visualize_snn_vs_cnn(scnn_compound, cnn_compound, 'spikingcnn_vs_cnn_compound.png')
```

```python
def visualize_limitation_cnn(li1, li2, li3, li4, path, color_idx=0):
    import numpy as np
    # x축에 해당하는 값 (인덱스 또는 다른 값)
    xticks=['20', '25', '30', '35']
    x = np.arange(len(xticks))
    bar_width=0.20

    # 그래프 그리기
    plt.figure(figsize=(10, 8))  # 그래프 크기 설정
    # 각 데이터 리스트에 대해 막대그래프 그리기 (x축 위치를 조금씩 옮김)

    colors = [
        ['#2066a8', '#3594cc', '#8cc5e3'],
        ['#a00000', '#c46666', '#d8a6a6'],
        ['#1f6f6f', '#54a1a1', '#9fc8c8']
    ]

    plt.bar(x - 1.5 * bar_width, li1, width=bar_width, label='Strong', color=colors[color_idx][0])  # 첫 번째 막대
    plt.bar(x - 0.5 * bar_width, li2, width=bar_width, label='Intermediate', color=colors[color_idx][1])  # 두 번째 막대
    plt.bar(x + 0.5 * bar_width, li3, width=bar_width, label='Weak', color=colors[color_idx][2])  # 세 번째 막대
    plt.bar(x + 1.5 * bar_width, li4, width=bar_width, label='SpikingCNN', color='#ea801c')  # 네 번째 막대

    # 그래프 제목 및 레이블 설정
    plt.xlabel('PSNR (dB)', fontsize=16)
    plt.ylabel('Accuracy (%)', fontsize=16)
    plt.xticks(x, xticks)
    
    # x축, y축 tick 크기 조정
    plt.tick_params(axis='x', labelsize=14)  # x축 tick 크기 설정
    plt.tick_params(axis='y', labelsize=14)  # y축 tick 크기 설정

    # 범례 추가
    plt.legend(fontsize=14, loc='upper center', bbox_to_anchor=(0.5, 1.1), ncol=4)

    plt.savefig(path)

    # 그래프 출력
    plt.show()
```

```python
very_strong_compound = [76.96, 90.28, 90.93, 89.84]  # 첫 번째 리스트
intermediate_compound = [69.77, 92.14, 94.97, 95.23]  # 두 번째 리스트
very_weak_compound = [56.50, 82.63, 91.07, 94.18]  # 두 번째 리스트
scnn_compound = [55.02, 87.72, 95.29, 98.42]  # 첫 번째 리스트

visualize_limitation_cnn(very_strong_compound, intermediate_compound, very_weak_compound, scnn_compound, 'limitation_noise_trained_cnn_compound.png', 0)
```

```python
very_strong_jitter = [78.32, 84.75, 88.70, 89.03]  # 첫 번째 리스트
intermediate_jitter = [89.30, 92.78, 94.95, 95.40]  # 두 번째 리스트
very_weak_jitter = [85.72, 91.03, 94.19, 94.26]  # 두 번째 리스트
scnn_jitter = [90.01, 98.00, 98.41, 98.51]

visualize_limitation_cnn(very_strong_jitter, intermediate_jitter, very_weak_jitter, scnn_jitter, 'limitation_noise_trained_cnn_jitter.png', 1)
```

```python
very_strong_uniform = [24.87, 70.05, 90.56, 89.74]  # 첫 번째 리스트
intermediate_uniform = [29.75, 77.16, 94.15, 95.23]  # 두 번째 리스트
very_weak_uniform = [36.39, 69.19, 88.99, 94.09]  # 두 번째 리스트
scnn_uniform = [9.74, 22.06, 89.82, 98.39]

visualize_limitation_cnn(very_strong_uniform, intermediate_uniform, very_weak_uniform, scnn_uniform, 'limitation_noise_trained_cnn_uniform.png', 2)
```
